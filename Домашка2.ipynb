Возьмем кусок сказки Лауры Нумерофф

*Если дать мышонку печенье, он попросит молока*

*Нальёшь ему молока — он захочет пить через соломин­ку*

*А когда всё выпьет, ему понадобится салфетка.*

С клавиатуры вводится, что мы дали мышонку: молоко, соломинку или салфетку. 

Выводится, что нужно мышонку дальше. 

В случае с салфеткой, считаем, что дальше он опять попросит стакан молока.

Ввод: Мышонку дали: молоко

Вывод: Мышонок попросил: соломинку

```python
a = "печенье"
b = "молоко"
c = "соломинку"
d = "салфетку"
word = input("Мышонку дали ")
if word == a:
  next_word = b
elif word == b:
  next_word = c
elif word == c:
  next_word = d
elif word == d:
  next_word = a
else: 
  next_word = "ошибка!"
print(f"Мышонок попосил {next_word}")

```

Дан день недели в виде числа (1 — понедельник, 2 — вторник и т. д.), выведите, является ли это рабочим днем ​​или выходным.

```python
day_n = int(input())
if (day_n >= 1 and  day_n <= 5):
  print("Будний день")  
elif (day_n >= 6 and  day_n <= 7):
  print("Выходной день") 
else:
  print("ошибка") 
```

Напишите программу, которая принимает ввод от пользователя и определяет, больше ли число 10, равно либо меньше

```python
num = int(input())
if num > 10:
  print(f"{num} > 10")  
elif num == 10:
  print(f"{num} == 10") 
else:
  print(f"{num} < 10") 
```

Реализуйте следующую программу: 

На экран выводится текст

Вы оказываетесь в странном лесу с тропинкой, разветвляющейся в двух направлениях. Один путь ведет к темному замку со зловещей репутацией, а другой ведет к таинственным руинам, которые, как считается, обладают магической силой.

Какой путь вы бы выбрали?

Игрок выбирает вариант ответа, в зависимости от этого меняется сюжет

```python
print("Вы оказываетесь в странном лесу с тропинкой, разветвляющейся в двух направлениях. Один путь ведет к темному замку со зловещей репутацией, а другой ведет к таинственным руинам, которые, как считается, обладают магической силой.") 
answer = input("Какой путь вы бы выбрали?")
word1 = "замок"
word2 = "руины"
if answer == word1:
  print("Вы оказываетесь к темному замку со зловещей репутацией")  
elif answer == word2:
  print("Вы оказываетесь к таинственным руинам, которые, как считается, обладают магической силой") 
else:
  print("Ничего не выбрали и остались стоять")
```
